```python
n=int(input())
def check_amstrong(b):
    c=str(b)
    d=len(c)
    result=0
    for i in c:
        digit=int(i)
        result+=digit**d
    if result==b:
        return'amstrong number'
    else:
        return'not an amstrong number'
for i in range(1,n+1):
    if check_amstrong(i)=='amstrong number':
        print(i)
```

```python
def is_prime(num):
    if num<1:
        return False
    for i in range (2,int(num**(1/2))+1):
        if num%i==0:
            return False
    return True
a=int(input())
if is_prime(a)==True:
    print(a,'is a prime number')
```
