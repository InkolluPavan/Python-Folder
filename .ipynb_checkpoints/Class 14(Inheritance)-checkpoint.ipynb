### Inheritance
- Simple Inheritance (Parent---->Child)
- Multi Level Inheritance (Grand Parent---->Parent---->Child)
- Muliple Inheritance((Parent1,Parent2)---->Child)
- Hierarchical Inhertiance(Single Parent----->(Child1,Child2,,,Child3))

```python
class Maruthi:
    company='Maruthi'
    @classmethod
    def comp_details(cls):
        return f'The Brand name is {cls.company}'
class Baleno(Maruthi):
    Model_name='Baleno'
    @classmethod
    def brand_name(cls):
        return f'The Model name is {cls.Model_name}'
mar_1=Maruthi()
print(mar_1.company)
print(mar_1.comp_details())

bal_1=Baleno()
print(bal_1.company) # It inherited the attribute company from the parent class 'Maruthi'
print(bal_1.comp_details()) # It inherited the attribute company_details from the parent class 'Maruthi'
print(bal_1.brand_name()) # It is the new attribute which is cretaed in the child class
```

```python
# Parent class is JNTUH
# Child class is TKR
class University:
    university='JNTU Hyderabad'
    @classmethod
    def uni_name(cls):
        return f'The University is {cls.university}'
class college(University):
    college='TKR'
    @classmethod
    def info(cls):
        return f'The University is {cls.university} and the College is {cls.college}'
uni_1=University()
print(uni_1.uni_name())
tkr=college()
print(tkr.info())
```

```python
#Single inheritance
class Indians:
    country_name='India'
    def __init__(self,name):
        self.name=name
    def name_info(self):
        return f'My name is {self.name}'
    @classmethod
    def country_details(cls):
        return f'My country is {cls.country_name}'
indian_1=Indians('Narendra Modi')
class Telangana(Indians):
    state_name='Telangana'
    def __init__(self,name,designation):
        Indians.__init__(self,name)
        self.designation=designation
    def full_info(self):
        return f'{self.name_info()} and {self.country_details()} and My state is {self.state_name} and My designation is {self.designation}'
tel_1=Telangana('Narendra Modi','DS')
print(tel_1.country_name)
print(tel_1.name)
print(tel_1.state_name)
print(tel_1.designation)
print(tel_1.full_info())
```

```python
class Axis:
    bank_name='Axis'
    @classmethod
    def bank_info(cls):
        return f'Bank name is {cls.bank_name}'
    #let's create instance variable
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance
    #Instance method
    def basic_info(self):
        return f'Hey {self.name},Your Bank Balance is {self.balance}.'
#let's create an instance of it
cust_1=Axis('Pavan',50000)
print(cust_1.bank_name)
print(cust_1.bank_info())
print(cust_1.basic_info())
class Lbnagar(Axis):
    branch_name='Lb Nagar'
    #Let's create an instance varibale account_type
    def __init__(self,name,balance,account_type):
        Axis.__init__(self,name,balance)
        self.account_type=account_type
    #let's create a new method to call out complete details using basic_info function created in parent class
    def complete_details(self):
        return f'{self.basic_info()}and your account type is {self.account_type}'
#let's create and instance of Lbnagar class
lb_1=Lbnagar('Pavan',50000,'Savings')
print(lb_1.complete_details())
```

### Multi-Level Inheritance

```python
#let's explore multi level inheritance 
#Grand parent---->Parent---->Child Class
#University ----->College---->Dept
class University:
    uni_name='JNTUH'
    @classmethod
    def university_name(cls):
        return f'Hey Welcome to {cls.uni_name} University.'
class College(University):
    col_name='TKR'
    @classmethod
    def college_name(cls):
        return f'Hey Welcome to {cls.col_name} college'
class Dept(College):
    dept_name="C.Sc"
    @classmethod
    def department_name(cls):
        return f"Hey Welcome to {cls.dept_name} department"
stu_1=Dept()
print(stu_1.university_name())
print(stu_1.college_name())
print(stu_1.department_name())
```

### Multiple Interitance
- (Parent_1)+(Parent_2)---->(Child)

```python
class Engine:
    engine_name='Maruti'
    @classmethod
    def eng_name(cls):
        return f'The Engine Belongs to {cls.engine_name},'
class Brand:
    brand_name='Toyota'
    @classmethod
    def Brand(cls):
        return f'The Brand is {cls.brand_name},'
class Glanza(Engine,Brand):
    model_name='Glanza'
    def com_details(cls):
        return f'{cls.eng_name()}{cls.Brand()} and the model is {cls.model_name}'
galnza_1=Glanza()
print(galnza_1.com_details())
```

```python
class Order:
    order_name = 'Iphone 17 Pro Max'
    @classmethod
    def ord_1(cls):
        return f' Hey your order {cls.order_name} is placed'
class Payment:
    payment_method='UPI'
    @classmethod
    def payment_1(cls):
        return f' The Payment was done in  {cls.payment_method} method'
class Delivery(Order,Payment):
    @classmethod
    def Delivery_1(cls):
        return f'{cls.ord_1()},{cls.payment_1()}, Hey your product {cls.order_name} is Delivered'
ord_1 = Delivery()
print(ord_1.Delivery_1())
```

```python
class JioCinema:
    name1='Jio Cinema'
    after_name='Jio'
    @classmethod
    def ott1(cls):
        return f'The Ott is {cls.name1}'
class HotStar:
    name2='Hotstar'
    @classmethod
    def ott2(cls):
        return f'The Ott is {cls.name2}'
class JioHotstar(JioCinema,HotStar):
    def ott(cls):
        return f'The New Ott is {cls.after_name}{cls.name2}'
jh=JioHotstar()
print(jh.ott1())
print(jh.ott2())
print('Both',jh.name1,'and',jh.name2,'are combined and')
print(jh.ott())
```

### Polymorphism

```python
class Strings:
    def add(self,a,b):
        self.a=a
        self.b=b
        return self.a+self.b
class Numeric:
    def add(self,x,y):
        self.x=x
        self.y=y
        return self.x+self.y
str_1=Strings()
print(str_1.add('Venkata ','Pavan'))
num_1=Numeric()
print(num_1.add(6,9))
```

```python
class UPI:
    def pay(self,amount):
        self.amount=amount
        return f'Hey, {self.amount} has been credited'
class CreditCard:
    def pay(self,amount):
        self.amount=amount
        return f'Hey, {self.amount} has been spent in your credit card.'
class BankAccount:
     def pay(self,amount):
        self.amount=amount
        return f'Hey, {self.amount} has been credited to your account.'

#Let's create an instance of UPI
upi_1=UPI()
print(upi_1.pay(5000)) #We have used pay method on the object of UPI class

#Let's create an instance of CreditCard
cred_1=CreditCard()
print(cred_1.pay(6000))

#Let's create an instance of BankAccount
bankacc_1 =BankAccount()
print(bankacc_1.pay(7000))
```

```python
class Dog:
    def voice(self,voice):
        self.voice=voice
        return f'{self.voice}'
class Cat:
     def voice(self,voice):
        self.voice=voice
        return f'{self.voice}'
class Bird:
     def voice(self,voice):
        self.voice=voice
        return f'{self.voice}'
dog_1 =Dog()
print(dog_1.voice('BOW BOW!'))

cat_1 =Cat()
print(cat_1.voice('MEOW MEOW!'))

bird_1 =Bird()
print(bird_1.voice('KUKOO KUKOO!'))
```

```python
# When an e-commerce platform shitfs a product it can use fedex,BHF,BlueDot,IndiaPost
# All Have different internal roles but platform calles a same method "shipment"
#Define a python program using ploymorphism where 
class fedex:
    def shipment(self,product):
        self.product=product
        return f'Your product "{self.product}" is dispatched by "FEDEX"'
class BHF:
        def shipment(self,product):
            self.product=product
            return f'Your product "{self.product}" is dispatched by "BHF"'
class BlueDot:
        def shipment(self,product):
            self.product=product
            return f'Your product "{self.product}" is dispatched by "BlueDOt"'
class IndianPost:
        def shipment(self,product):
            self.product=product
            return f'Your product "{self.product}" is dispatched by "IndianPost"'
ord_1=fedex()
print(ord_1.shipment('IPHONE 17 PRO MAX'))

ord_2=BHF()
print(ord_2.shipment('SAMSUNG S24 ULTRA'))

ord_3=BlueDot()
print(ord_3.shipment('ONEPLUS 15'))

ord_4=IndianPost()
print(ord_4.shipment('IQOO 15'))
```

```python
class Car:
    def __init__(self,speed,color):
        self.speed=speed
        self.color=color
    def set_speed(self,value):
        self.speed=value
    #Getter for the atribute speed
    def get_speed(self):
        return self.speed

ford = Car(200,'Red')
Chev = Car(180,'Blue')
Audi = Car(190,'Black')
ford.set_speed(400)
ford.speed=500
print(ford.get_speed())
```

```python
class Car:
    def __init__(self,speed,color):
        self.__speed=speed
        self.__color=color
    #Setter for the attribute speed
    def set_speed(self,value):
        self.__speed=value
    #Getter for the atribute speed
    def get_speed(self):
        return self.__speed
    #Setter for the attribute color
    def set_color(self,value):
        self.__color=value
    #Getter for the atribute color
    def get_color(self):
        return self.__color
ford = Car(200,'Red')
Chev = Car(180,'Blue')
Audi = Car(190,'Black')
ford.set_speed(400)
ford.set_color('Red')
print(ford.get_speed())
print(ford.get_color())
```

```python
class Rectangle:
    def __init__(self,length,width):
        self.__length=length
        self.__width=width
    def set_length(self,value):
        self.__length=value
    def get_length(self):
        return self.__length
    def set_width(self,value):
        self.__width=value
    def get_width(self):
        return self.__width
    def area(self):
        return self.__length * self.__width 
rect_1=Rectangle(60,20)
print(rect_1.get_length())
print(rect_1.get_width())
rect_1.area()
```
