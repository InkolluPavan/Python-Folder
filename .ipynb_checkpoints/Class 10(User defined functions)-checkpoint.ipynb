```python
# write a user defined function to check wheater given word is palindrome or not
def palindrome(x):
    if x.lower()==x[::-1].lower():
        return True
    else:
        return False
palindrome("madam")        
```

```python
# write a function that takes a list of integers and return a new list containing only even numbers from original list
list_1 = [ 1,2,3,4,5,6,7]
def even(list_1):
    new_list=[]
    for x in list_1:
        if x%2==0:
            print(x)
            return new_list.append(x)
    return new_list   
even(list_1)                   
```
