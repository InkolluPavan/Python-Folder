```python
#lets create a dictionary with name salary savings and designations
#Each key should contain 5 values
import pandas as pd
em=pd.DataFrame({'Name':['Pavan','Malika','Tanuja','Shamu','Divya'],
        'Designation':['Manager','Team lead','Developer','Tester','Customer Support'],
        'Salary':[50000,45000,40000,25000,20000],
        'Savings':[25000,15000,10000,5000,2500]})
em
```

```python
#now fetch designation and salary coloums at a time
em[['Designation','Salary']]
```

```python
#Using Salary and Saving coloums, Lets create a new column Expenditure
#Expenditure =Salary -Savings
em['Expenditure'] =em['Salary']-em['Savings']
em
```

```python
#find what percentage of salary is going into savings and expenditure and add those new columns
em['Saving_percentage']=round(em['Savings']/em['Salary']*100,2)
em
```

```python
em['Expenditure_percentage']=round(100-em['Saving_percentage'],2)
em
```

```python
em['First_Name']=em["Name"]
em
```

```python
em.drop('First_Name',axis=1,inplace=True)
em
```

```python
em=em.rename(columns={'Name':'First_Name'})
em
```

```python
emp=em
emp
```

```python
emp.drop(3,axis=0,inplace=True)
emp
```

```python
emp.set_index('First_Name',inplace=True)
emp
```

```python
#create a new dataFrame
#First_Name
#Last_Name
stu_details=pd.DataFrame({
    'First_name':['Aravind','Shubhan','Abdul','Venkata','Saketh'],
    'Last_name':['Reddy','Shaik','Sumain','Pavan','Aditya'],
    'Qualification':['MBA','B.E','B.Tech','B.Tech','B.Tech']
})
stu_details
```

```python
#create a new coloumn 'Full_Name' by using 'First_Name' and 'Last_Name'
stu_details['Full_Name']=stu_details['First_name']+' '+stu_details['Last_name']
stu_details
```

```python
#lets Change The name "Qualification" to "Education"
stu_details=stu_details.rename(columns={'Qualification':'Eductaion'})
stu_details
```
