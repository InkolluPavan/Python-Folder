```python
#Replace the second element "banana" with "mango" in the list: fruits = ["apple", "banana", "orange"].
fruits = ["apple", "banana", "orange"]
for i in fruits:
    if i=="banana":
        fruits.remove('banana')
        fruits.insert(1,'mango')
print(fruits)
```
