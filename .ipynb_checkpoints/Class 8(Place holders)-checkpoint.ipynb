### %Formating
- %s -String Placeholder
- %d -Integer Placeholder
- %f -Float Placeholder

```python
Name = input("Name of the Student is:")
Age = int(input("Age of Student is:"))
Course = input("Course Enrolled for is:")
print("Hey %s , Welcome to %s program offered by Unicus Learning." %(Name, Course) )
print("Hey %s and your age is %d." %(Name, Age))
```

```python
# Write Python Program that takes length and width of a rectangle as dynamic inputs and prints the Area and Perimeter of rectangle
# Area = Length*width
# Perimeter = 2*(Length + Width)
# Make sure to use % formatting for placeholders
length=float(input("Enter the length : "))
width=float(input("Enter the width : "))
area= length*width
perimeter=2*(length+width)
print("Area of the rectangle with length %f and width %f is %d and the perimeter is %d." %(length, width, area, perimeter) )
```

```python
# Write Python code to find the the Area and Circumference of a circle when the diameter of the circle is given
dia=float(input("Enter the diameter: "))
rad=dia/2
ar=(22/7)*(rad**2)
per=2*(22/7)*rad
print('The radius is %f and the area is %f and the perimeter is %f'%(rad,ar,per))
```

### .format()

```python
name=input("Enter your name : ")
percent=float(input("enter your percentage : "))
print('hey {} ,Your percentage is {:.2f}'.format(name,percent))
```

```python
name=input("Enter your name : ")
percent=float(input("enter your percentage : "))
print('hey {0} ,Your percentage is {1}'.format(name,percent))
```

### f String method

```python
name=input("Enter your name: ")
percentage=float(input("Enter your percentage"))
print(f'hey {name}, Your percentage is {percentage:.2f}')
```
