```python
#let's Create a class and name its Students 
class Students:
    university ='Unicus Learning'
    def __init__(self,name,age,course):
        self.name=name
        self.age=age
        self.course=course

#let's fetch the class attributes using the class name itself
print(Students.university)

#let's create an object
stu_1=Students('Pavan',23,'AIML')
stu_2=Students('Sravani',22,'DS')
#Lets assign an attribute, gender to stu_1,stu_2
stu_1.gender = 'Male'# Here the 'gender' is the dynamic attribute which is assigned to only a particular object
stu_2.gender = 'Female'# Here the 'gender' is the dynamic attribute which is assigned to only a particular object
print(stu_1.university)# We are calling the class attributes using object name
print(stu_1.name)# We are fetching instance attribute using object name
print(stu_1.age)# We are fetching instance attribute using object name
print(stu_1.course)# We are fetching instance attribute using object name
print(stu_1.gender)
print(stu_2.university)
print(stu_2.name)
print(stu_2.age)
print(stu_2.course)
print(stu_2.gender)
```

- Class attributes can be called using class and objects
- Instance attributes can be called only using the objects/instances
- Dynamic attributes can be called only using the instances/objects to which we have assigned it

```python
class Banks:
    bank_name='Andhra Bank'
    def __init__(self,name,branch,city):
        self.name=name
        self.branch=branch
        self.city=city
    def welcome(self):
        return f"Hey {self.name} Welcome to {self.branch} branch of Andhra Bank."
    @classmethod
    def change_name(cls,new_name):
        cls.bank_name=new_name
        return cls.bank_name
def greetings(self):
    return f'Hey {self.name} ,{self.bank_name} Wishing you Happy Sankrathi'
def hoildays(self):
    return f"'Hey {self.name}, Our {self.bank_name}'s {self.branch} branch will be closed from 24th Dec to 27th Dec'"
Banks.greetings =greetings
acc_1=Banks('Bob','Ladki-Ka-Pool','Hyderbad')
print(Banks.bank_name)
print(acc_1.name)
print(acc_1.branch)
print(acc_1.city)
print(acc_1.welcome())
acc_1.change_name('Union Bank of India')
print(Banks.bank_name)
print(acc_1.greetings())

#lets assign the dynamic method holidays() to acc_1
import types
acc_1.hoildays =types.MethodType(hoildays,acc_1)
print(acc_1.hoildays())
```

```python
# Let's Create  a class of RTC
# Under that create a class variable named 'corp_name' and assign 'APSRTC' to that
# Create two instance attributes 'name','start_point' and 'end_point'
# Create a class method to change the name of {corp_name} to 'TGSRTC'
# As soon as someone books a ticket, it should send them a Message Saying that 'Welcome to {corp_name}'
# Create an instance method to Print -Hey {Name}, Your Ticket from {start_point} to {end_point} is successfully booked
# Create two instances of RTC and call all attributes and methods on those
class RTC:
    corp_name='APSRTC'
    def __init__(self,name,start_point,end_point):
        self.name=name
        self.start_point=start_point
        self.end_point=end_point
    def changed_name(cor,new_name):
        cor.corp_name=new_name
    def journey(self):
        return f'Hey {self.name} Your Ticket from {self.start_point} to {self.end_point} is succesfully booked'
def welcome(self):
    return f'Hey {self.name} Welcome to {self.corp_name}'
RTC.welcome =welcome 
def cancellation(self):
    return f'Hey {self.name}, Your jounery from {self.start_point} to {self.end_point} is cancelled due to faliure in payment'
user_1= RTC('Pavan','Ongole','Hyderabad')
user_2= RTC('Sravani','Addanki','Hyderabad')
print(user_1.welcome())
user_1.changed_name('TGSRTC')
user_1.changed_name('TGSRTC')
print(user_1.welcome())
print(user_1.journey())
print(user_2.journey())
import types
user_2.cancellation=types.MethodType(cancellation,user_2)
print(user_2.cancellation())
```

```python
# You are buliding a small system to represent an online course
# 1.Create a class named 'Course'
# 2.Add a class attributes Platform  and assign  a value called 'UNICUS LEARNING'
# 3.Add an instance attributes 'Name' ,'course_name' and 'Price'
# 4.Add an instance method and name it as show_details
# 5. Create one object of the class then dynamically add an attribute instructor
# 6. Then dynamically add a method and name it as apply_discount ,This method shoul reduce the course price based on the discount
# 7 Then call all the methods
class Course:
    Platform='UNICUS LEARNING'
    instructor=''
    def __init__(self,name,course_name,price):
        self.name=name
        self.course_name=course_name
        self.price=price
    def show_details(self):
        return f'Hey! {self.name} Welcome to {self.Platform} ,Your instructor is {self.instructor},You enrolled to {self.course_name} Course and your fee is ₹{self.price}/-'
@classmethod
def assign_instructor(ins,new_name):
    ins.instructor=new_name
Course.assign_instructor=assign_instructor
def discount(self,des_price):
    self.des_price=des_price
    final_price= int(self.price *(100-des_price)/100)
    return f'You got {self.des_price}% discount and your final price is ₹{final_price}/-'
stu1=Course('Srujan','Data Science',50000)
stu1.assign_instructor('Aravind')
print(stu1.show_details())
import types
stu1.discount=types.MethodType(discount,stu1)
print(stu1.discount(30))
```
