```python
def sq_sqrt(a):
    print('The square of number is:' ,a**2)
    print('The sqroot of number is:',a**(1/2))
d=int(input("Enter a value : "))
sq_sqrt(d)
```

```python
#write  a function to find the factorial of a number
def fact(n):
    fact=1
    for i in range(1,n+1):
        fact*=i
    return fact
n=int(input("Enter any number:"))
print('The factorial of the number',n,'is :',fact(n))
```

```python
def eve_odd(n):
    if n%2==0:
        return True
    return False
n=int(input("Enter a number: "))
if eve_odd(n)==True:
    print('Even Number')
else:
    print('Odd Number')
```

```python
#write a function to covert celsius to faren heat
celcius= float(input("Enter the temperature in celcius"))
def cel_fheat(n):
    f=n*(9/5)+32
    print("The farenheat of the value",n,"is: ",round(f))
cel_fheat(celcius)
```

```python
def avg(a,b,c,d):
    sum=(a+b+c+d)/4
    print('The Average of Given numbers is: ',sum)
a=int(input("Enter 1st number"))
b=int(input("Enter 2nd number"))
c=int(input("Enter 3rd number"))
d=int(input("Enter 4th number"))
avg(a,b,c,d)
```
