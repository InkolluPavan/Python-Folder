- Tuples are imutable
- tuples only have two attributes like count() and index()

```python
def isprime(n):
    if n<=1:
        return False
    for i in range(2,int(n**(1/2))+1):
        if n%i==0:
            return False
    return True
lis_2=[x for x in range(1,101) if isprime(x)==True]
print(lis_2)
print(len(lis_2))
```
