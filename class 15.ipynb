```python
def isprime(n):
    if n<2:
        return False
    for i in range(2, int(n**(1/2)+1)):
        if n%i==0:
            return False
    return True
prime=[]
n= int(input('Enter the staring number'))
m=int(input('Enter the ending number'))
for i in range(n,m+1):
    if isprime(i):
        prime.append(i)
print(f"Found {len(prime)} primes: {prime}")
```
