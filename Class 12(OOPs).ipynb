```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def intro(self):
        return f"My Name is {self.name} and My age is {self.age}."

#Lets create an instance to Person class 
p1=Person('Pavan',23)
print(p1.age)
print(p1.name)
print(p1.intro())
p2=Person('Sarath',29)
print(p2.age)
print(p2.name)
print(p2.intro())
```

```python
#create a class of "cars" 
#Assign two attributes brand and model
#Define a method  and name it as "about"
#The brand name is {Brand} and the mode is {model}
#Create two insttances of cars
#One with brand ='Maruthi' and model='Swift'
class Car:
    def __init__(self,brand,model):
        self.brand=brand
        self.model=model
    def about(self):
        return f"The Brand name is {self.brand} and the model is {self.model}."
c1=Car('Maruthi','Swift')
print(c1.about())
c2=Car('Tata','Curv')
print(c2.about())
```

```python
# Create a class and name it bank account
# Assign it two variable name and balance 
# Create a method to print the name and opening balance 
# Create a method called depoist which would print the ammount deposited and updated balance in the account
# Create a instance of it and apply deposit menthod on it
class Bank:
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance
    def user_info(self):
        return f'Hello {self.name} .Welcome to you bank and Your opening balance is {self.balance}'
    def deposit(self,amount):
        self.balance+=amount
        return f"The Amount added is {amount} and The updated balance is {self.balance}"
    def withdrawl(self,amount):
        if self.balance  >= amount:
            self.balance -= amount
            return f"Amount withdrawl is {amount} and the current balance is {self.balance}"
        else:
            return f"Amount tried to withdrawn is {amount} and you have inssufficnent funds"
user1= Bank("Pavan",10000)
print(user1.user_info())
print(user1.deposit(3000))
print(user1.withdrawl(2000))
```

```python
# Create a class and name it as Unicus learning
# Give 5 attributes Name,Age ,Qualification ,Course and Fee
# Create a function name it as Welcome which should print the name and course enrolled for
# Create a instance of it and print the attributs and method applied on it
#Create a function to fetch the fee pending when a cretain amount is paid
class Unicus_Learning:
    def __init__(self,name,age,qualification,course,fee):
        self.name=name
        self.age=age
        self.qualification=qualification
        self.course=course
        self.fee=fee
    def welcome(self):
        return f"Hey! {self.name} Welcome to Unicus Learning . Congratulations! You have enrolled To {self.course} Course"
    def pending_fee(self,amount):
        pending =self.fee-amount
        return f"You had paid the amount of ${amount} and The Pending fee is ${pending}"
stu1=Unicus_Learning('Pavan',23,'B-Tech','Data Analytics',55000)
stu2=Unicus_Learning('Sravan',23,'B-Com','Data Enigneering',55000)
print(stu1.welcome())
print(stu1.pending_fee(40000))
```
