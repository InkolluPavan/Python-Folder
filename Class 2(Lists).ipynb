```python
# Assume that on the first day 3 students enrolled for data science course
# We want to save that information into some python objects
students =['Vinay','Heba','Pavan']
print("Data Type of students Object is : ",type(students))
print(students)
```

```python
# on the next day , Saketh took the admission
# Now we have to add 'Saketh ' to the list of 'students'
students.append('Saketh')
print(students)
```

```python
#Next we have to add shuban and sumair to the 'students' list
students.append('Shuban','Sumair')
print(students)
```

```python
students.append(['Subhan','Sumair'])
print(students)
```

```python
students.pop() #Using pop() function we can get rid odd the last or recent (?)
print(students)
```

```python
students.extend(['Subhan','Sumair']) #Using extend function, we can add multiple elements to the list at a time
print(students)
#extended two times
```

```python
print(students)
```

```python
# Actually 'Vijay' was the first person to take the admission but we forgot it
# Using insert fuction and proper index add 'Vijay' as the first element in the list
students.insert(0,'Vijay')
print(students)
```

```python
# But the Vijay mentioned taht we havr to add his full name 
# So we have to first  get rid off the element 'Vijay' from the list of students
students.pop()
print(students)
```

```python
#by mistake the  element sumair is gone ,let's add sumair to the list
students.append('Sumair')
print(students)
```

```python
#lets remove the name 'Vijay' using remove function
students.remove('Vijay')
print(students)
```

```python
#now lets add 'Vijay krishna' to the list as first index
students.insert(0,'Vijay Krishna')
print(students)
```

```python
#Copy function create a copy of the list
student_copy =students.copy()
print(student_copy)
```

```python
#Count function would help us finding the count of the elements in the list
students.count('Pavan')
```

```python
#Index function would fetch the index of an element in the list
students.index('Vinay')
```

```python
students.sort()
print(students)
```

```python
students.reverse()
print(students)
```

### Differnt functions applicable on lists
- append() function will help in adding one new elements to the list at a time
- pop() function will help in removing the last index of the list
- remove() function will help in removing particular element in the list
- extend() function will help in adding multiple elements to the list
- insert() function will help in adding to insert in particular index
- copy() function will help in making a copy of the list
- count() function will help in counting the element count in the list
- sort() function will help in sorting the list
- reverse() function will help in reversing the list
