```python
#create a list of any 5 states
states=['Andhra Pradesh','Karnataka','Maharastra','Himachal Pradesh','Madya Pradesh']
print(states)
```

```python
# Using a for loop, convert every element of 'states' into lower case and store it into another list
lower_state=[]
for i in states:
    i=i.lower()
    lower_state.append(i)
print(lower_state)
```

```python
states2=[x.lower() for x in states]
print(states2)
```

```python
# Using for loop create multiplication table of 7
n=int(input("Enter the number :"))
for i in range(1,11):
    print(n,'*',i,'=',n*i)
```

```python
#lets count thr vowels in a string
a=input()
word='aeiou'
count=0
for i in a.lower():
    if i in word:
        count+=1
print(count)
```

```python
#write python code to find sum of number in given list
n=int(input("Enter the size of list : "))
int_lis=[]
for i in range(1,n+1):
    a=int(input())
    int_lis.append(a)
total=0
for i in int_lis:
    total+=i
print(total)
```

```python
def count(a):
    a=a.lower()
    word='aeiou'
    vowelcount=0
    consonantcount=0
    for i in a:
        if i in word:
            vowelcount+=1
        else:
            consonantcount+=1
    print('Vowel count:',vowelcount)
    print('Consonant count:',consonantcount)
a=input()
count(a)
```
