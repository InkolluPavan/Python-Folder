#### Sets
- Sets don't allow Duplicates 
- Sets are always with sorting

```python
set_1={9,3,5,7,8,1,2,2,-3}
print(type(set_1))
set_1
```

```python
#create another set with a few common elements from set_1 and name it set_2
set_2={-1,-3,5,-7,5,3,7,7,8,9,3,4}
print(type(set_2))
set_2
```

```python
set_1.add(14)#here add function is used to add a new element to the set
set_1
```

```python
set_1.intersection(set_2)
```

#### intersection update would replace the first set common elements an second set would remain same
#### Second set would remain same

```python
set_1.intersection_update(set_2)
set_1
```

```python
set_3={81,2,3,4,5,5,61,3,8,4}
set_4={6,2,4,67,2,5,1,43,6,8,}
print(set_3)
print(set_4)
```

```python
set_3.difference(set_4)
set_4
```

#### Union

```python
set_3.union(set_4)#combines two sets
```

#### Clear

```python
set_3.clear()#Clear the set elements
```

```python
set_3
```

#### Pop

```python
set_4.pop() #pops the first element in the set
```

```python
set_4
```

#### Copy

```python
set_4=set_1.copy() #Copy the set to another set
set_4
```

#### Update

```python
set_4.update(set_2) #combines the two sets and update it in one set
set_2
```

#### isdisjoint

```python
set_4.isdisjoint(set_1)# return true is there is no common element in them
```

#### issubset

```python
set_4.issubset(set_1) #return true if the set_1 is a subset of the set_4
```

#### is superset

```python
set_4.issuperset(set_1) # return true if the set_1 is a super set of set_4
```
