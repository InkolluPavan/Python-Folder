```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. NumPy: Generate raw math data (100 random numbers representing sales)
raw_data = np.random.normal(500, 100, 100) 

# 2. Pandas: Organize that raw math into a labeled "Blueprint" (DataFrame)
df = pd.DataFrame(raw_data, columns=['Sales_Amount'])

# 3. Matplotlib: Create the "Window" to see the data (Visualization)
plt.hist(df['Sales_Amount'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Sales')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd

# 1. Load & Prepare Data (Pandas)
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target # The "Answer" we want to predict

# 2. Train-Test Split (Standard practice in your JD)
# We reserve 20% of data to test if the model actually learned
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)

# 3. Build & Train the Model (Scikit-learn)
model = RandomForestClassifier(n_estimators=100) # Creating the "Engine"
model.fit(X_train, y_train)                      # "Learning" from the training data

# 4. Make Predictions & Evaluate (Statistics)
predictions = model.predict(X_test)
print(f"Model Accuracy: {accuracy_score(y_test, predictions) * 100}%")
print("\nDetailed Report:\n", classification_report(y_test, predictions))
```
